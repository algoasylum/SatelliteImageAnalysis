```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
import seaborn as sns
sns.set(style='white')
pio.templates.default = "none"
```

```python
df=pd.read_csv('ndvi_all.csv')
del df['Unnamed: 0']
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d/%m/%Y')
df.head(5)
df['NDVI']=df['NDVI'].astype('float')
```

```python
fig = px.line(df,'Date','NDVI')
fig.show()
```

```python
sns.distplot(df['NDVI'])
```

```python
df['year']=df['Date'].dt.year
df5 = df[df.year < 2006]
df10 = df[(df.year >= 2006) & (df.year < 2011)]
df15 = df[(df.year >= 2011) & (df.year < 2015)]
df20 = df[df.year >= 2016]

df0010 = df[df.year < 2011]
df1120 =df[df.year >= 2011]
```

```python
f, ax = plt.subplots(1, 1)
sns.distplot(df5['NDVI'])
sns.distplot(df10['NDVI'])
sns.distplot(df15['NDVI'])
sns.distplot(df20['NDVI'])
plt.legend(labels=['00-05','06-10','10-15','15-20'])
plt.show()
sns.distplot(df0010['NDVI'])
sns.distplot(df1120['NDVI'])
plt.legend(labels=['00-10','11-20'])
plt.show()
```

```python
f, axes = plt.subplots(1, 4, figsize=(18, 4), sharex=True)
sns.distplot(df5['NDVI'],ax=axes[0],color='b')
axes[0].legend(labels=['00-05'])
sns.distplot(df10['NDVI'],ax=axes[1],color='r')
axes[1].legend(labels=['06-10'])
sns.distplot(df15['NDVI'],ax=axes[2],color='g')
axes[2].legend(labels=['11-15'])
sns.distplot(df20['NDVI'],ax=axes[3],color='m')
axes[3].legend(labels=['15-20'])
plt.setp(axes, yticks=[])
```
