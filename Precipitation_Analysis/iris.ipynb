```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import seaborn as sns
from sklearn import datasets
```

```python
iris = datasets.load_iris()

X = iris.data[:, :1]  # we only take the first two features.
X.shape
```

```python
x=np.array_split(X, 10)
```

```python
vec=[]
mat = []
for i in range(0,10):
    vec.append(x[i].mean())
    vec.append(x[i].std())
    xq = pd.DataFrame(x[i])
    xq.plot.hist()
    mat.append(vec)
    vec=[]
mat = np.asarray(mat)
mat
```

```python
matt =[]
for i in range(0,9):
    matt.append(mat[i+1] - mat[i])
matt
```

```python
dist = pd.DataFrame(matt)
cm = sns.light_palette("blue", as_cmap=True) 
dist.style.background_gradient(cmap=cm).set_precision(10)
```
