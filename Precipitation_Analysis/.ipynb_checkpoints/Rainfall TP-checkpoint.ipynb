```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.figure_factory as ff
from scipy import fftpack
import datetime
```

```python
daily_data = pd.read_csv('Precipitation data(csv)\preci_daily.csv')
daily_data.head()
#daily_data.loc[(daily_data['date'].between('2001-01-01', '2001-12-31', inclusive=True))]
```

```python
dates = [('2001-01-01','2001-12-31')]
for i in range(2,20):
    if i < 10:
        dates.append((dates[0][0][:3]+str(i)+dates[0][0][4:],dates[0][1][:3]+str(i)+dates[0][1][4:]))
    else:
        dates.append((dates[0][0][:2]+str(i)+dates[0][0][4:],dates[0][1][:2]+str(i)+dates[0][1][4:]))
dates
```

```python
yf = []
N = []
tp = []
ampli = []
for i,j in dates:
    N.append(daily_data.loc[(daily_data['date'].between(i,j, inclusive=True))].shape[0])
    sr = 1
    y = np.array(daily_data.loc[(daily_data['date'].between(i,j, inclusive=True))]['value'])
    yf.append(fftpack.fft(y))
    #T = 1.0 / N
    #x = np.linspace(0.0, N*T, N)
for i in range(len(N)):
    freq = np.fft.fftfreq(N[i])
    tp.append(1/freq)
    ampli.append(2.0/(N[i]) * np.abs(yf[i][:N[i]//2]))
    #ampli_tp = pd.DataFrame(zip(2.0/N * np.abs(yf[:N//2]), tp),columns=['Ampli', 'Time period'])
    #xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))
    #fig, ax = plt.subplots()
    #ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
    #plt.show()
```

```python
ampli_tp = pd.DataFrame(zip(2.0/N * np.abs(yf[:N//2]), tp),columns=['Ampli', 'Time period'])
ampli_tp
```

```python
sorted_data = ampli_tp.sort_values(by='Ampli', ascending=False)
sorted_data.drop([0], inplace = True) 
sorted_data = sorted_data.iloc[0:50]
```

```python
plt.figure(figsize=(16, 6))
k = []
for i in range(0, 750, 2):
    k.append(i)
sns.distplot(sorted_data['Time period'], bins = k, kde = False)
plt.xlim(0,50)
```
