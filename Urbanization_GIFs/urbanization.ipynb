```python
import geemap
```

```python
Map = geemap.Map(center=[20.5937, 78.9629])
Map
```

```python
import ee
urban_data = ee.ImageCollection("JRC/GHSL/P2016/SMOD_POP_GLOBE_V1") \
  .filterDate('1975-01-01', '2015-12-31')
Map.addLayer(urban_data, {'min' : 0, 'max' : 3, 'palette': ['white', 'yellow', 'blue', 'red']}, "Urbanization")
```

```python
count = urban_data.size()
print('Count: ', str(count.getInfo())+'\n')
```

```python
dates = ee.List(urban_data.aggregate_array("system:time_start"))
m = dates.map(lambda x:(ee.Date(x)).format('D-M-Y'))
date_list = m.getInfo()
date_list
```

```python
country_boundaries = ee.FeatureCollection("FAO/GAUL/2015/level2")
Map.addLayer(country_boundaries, {}, "District Boundaries")
```

```python
# Define arguments for animation function parameters.
videoArgs = {
  'dimensions': 500,
  'region': ee.FeatureCollection(Map.draw_last_feature),
  'framesPerSecond': 0.5 ,
  'min': 0.0,
  'max': 3.0,
  'palette': ['white', 'blue', 'yellow', 'red']
  #'maxPixels': 59636736,
}
```
