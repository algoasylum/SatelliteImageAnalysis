```python
import pandas as pd
```

```python
df = pd.read_csv("preci_all_delhi.csv")
df.info()
df
```

```python
df["system:index"] = pd.to_datetime(df["system:index"],format = '%Y%m%d%H%M%S')
#dff = df["system:index"].groupby(df["system:index"].dt.day).sum()
df = df.resample('D', on='system:index').sum()
```

```python
df
```

```python
del df['date']
del df['.geo']
```

```python
df=df.reset_index()
df.columns = ["date","value"]
df
```
