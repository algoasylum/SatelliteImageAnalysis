```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LogNorm
from sklearn import mixture
```

```python
value1 = np.random.normal(loc=5,scale=25,size=10000)
value2 =  np.random.normal(loc=100,scale=50,size=10000)
value3 =  np.random.normal(loc=500,scale=75,size=10000)
#value4 =  np.random.normal(loc=1000,scale=25,size=10)
soda = np.concatenate((value1,value2,value3,value4))
sns.distplot(value1)
sns.distplot(value2)
sns.distplot(value3)
```

```python
sns.distplot(soda,bins=1000)
```

```python
clf = mixture.GaussianMixture(n_components=3, covariance_type='full')
soda = soda.reshape(-1,1)
clf.fit(soda)
```

```python
print(clf.weights_)
print(clf.means_)
#print(clf.covariances_)
```
