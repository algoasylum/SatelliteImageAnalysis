```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import mixture
sns.set(style='whitegrid')
sns.set_palette('bright')
import warnings
warnings.filterwarnings('ignore')
import collections
from scipy.spatial import distance
from scipy.stats import norm
from scipy import stats
from scipy.stats import wasserstein_distance
```

```python
df = pd.read_csv('preci_alt.csv')
df.columns = ['date','value']
df.date = pd.to_datetime(df.date,format='%b %d, %Y')
```

```python
df2019 = df[(df.date.dt.year == 2019)]
df2018 = df[(df.date.dt.year == 2018)]
df2017 = df[(df.date.dt.year == 2017)]
df2016 = df[(df.date.dt.year == 2016)]
df2019
```

```python
df19 = df[df.date.dt.year == 2019]
fig, ax = plt.subplots(12,1, figsize=(18,25),sharey= True)
for i in range(0,12):
    dff = df19[df19.date.dt.month == i+1]
    sns.lineplot(dff.date,dff.value,ax = ax[i])
plt.show()
```

```python
minn=[]
maxx=[]
means=[]
df19 = df[df.date.dt.year == 2019]
for i in range(0,12):
    dff = df19[df19.date.dt.month == i+1]
    minn.append(dff.value.min())
    maxx.append(dff.value.max())
    means.append(dff.value.mean())
dmc = pd.DataFrame(zip(minn,maxx,means),columns=['min','max','means'])
dfw = pd.DataFrame(dmc)
cm = sns.light_palette("green", as_cmap=True)
s1 = dfw.style.background_gradient(cmap=cm)
minn=[]
maxx=[]
means=[]
df19 = df[df.date.dt.year == 2018]
for i in range(0,12):
    dff = df19[df19.date.dt.month == i+1]
    minn.append(dff.value.min())
    maxx.append(dff.value.max())
    means.append(dff.value.mean())
dmc = pd.DataFrame(zip(minn,maxx,means),columns=['min','max','means'])
dfw = pd.DataFrame(dmc)
cm = sns.light_palette("green", as_cmap=True)
s2= dfw.style.background_gradient(cmap=cm)
```
